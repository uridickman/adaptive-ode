```python
from solver.jit_functions import *
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(1234)
x = np.sort(np.random.random(6))
y = np.sort(np.random.random(6))
```

```python
DD = DD_table(x,y,order=len(x))
pd.DataFrame(DD)
```

```python
def interpolate(x_data, dd, x_query):
    _, n = dd.shape
    coeffs = dd[0, :]

    out = coeffs[0]
    mult_x = 1.0

    for k in range(1, n):
        mult_x *= (x_query - x_data[k - 1])
        out += coeffs[k] * mult_x

    return out
```

```python
plt.scatter(x,y)
xq = np.linspace(0.2,0.8,100)
plt.plot(xq,interpolate(x,DD,xq))
```
